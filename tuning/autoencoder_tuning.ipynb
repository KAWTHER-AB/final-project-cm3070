````python
### **1. Autoencoder Hyperparameter Tuning ‚Äì Summary** 

# Setup project path 
import sys, os, time
project_root = os.path.abspath(os.path.join(os.getcwd(), ".."))
if project_root not in sys.path:
    sys.path.append(project_root)

# --- Imports ---
import optuna
import pandas as pd
from tqdm import tqdm
from preprocessing.data_loader import load_data, preprocess_data, split_data, scale_data
from training.autoencoder_trainer import train_autoencoder, evaluate_autoencoder
from utils.style_utils import styled_print

# Data Prep
df = preprocess_data(load_data())
X_train, X_val, X_test, y_train, y_val, y_test = split_data(df)
X_train_scaled, X_val_scaled, X_test_scaled, scaler = scale_data(X_train, X_val, X_test)
X_train_0 = X_train_scaled[y_train == 0]
input_dim = X_train_scaled.shape[1]
epochs = 50

# --- Tuning Storage 
tuning_results = []
trial_durations = []
n_trials = 20

# --- Progress bar 
pbar = tqdm(total=n_trials, desc="üîç Bayesian Tuning Progress")

# Optuna Objective 
def objective(trial):
    start = time.time()

    enc_dim = trial.suggest_categorical('encoding_dim', [16, 32, 64])
    dropout = trial.suggest_float('dropout', 0.1, 0.5)
    batch_size = trial.suggest_categorical('batch_size', [128, 256])
    noise_std = trial.suggest_float('noise_std', 0.0, 0.2)

    model, _ = train_autoencoder(
        X_train_0, X_val_scaled,
        input_dim=input_dim,
        encoding_dim=enc_dim,
        dropout=dropout,
        noise_std=noise_std,
        batch_size=batch_size,
        epochs=epochs
    )

    metrics, _ = evaluate_autoencoder(model, X_val_scaled, y_val, visualize=False)  # validation set only

    # ‚è± Timing
    duration = round(time.time() - start, 2)
    trial_durations.append(duration)
    tuning_results.append({**trial.params, **metrics})

    # Progress bar info
    avg_time = sum(trial_durations) / len(trial_durations)
    remaining = avg_time * (n_trials - len(trial_durations))
    pbar.set_postfix_str(f"Est. time left: {remaining/60:.1f} min | Last F1: {metrics['f1_score']:.4f}")
    pbar.update(1)

    return metrics["pr_auc"]

# --- Run study ---
study = optuna.create_study(direction='maximize')
study.optimize(objective, n_trials=n_trials)
pbar.close()


styled_print("‚úÖ Tuning complete.")
````

````python
# --- Save best config ---
styled_print("üèÜ Best Autoencoder Config:")
print(study.best_params)

# --- Retrain on train+val (normal samples) ---
from numpy import vstack
X_trainval = vstack([X_train_0, X_val_scaled[y_val == 0]])

# ‚úÖ Save to tuning_results/
os.makedirs("tuning_results", exist_ok=True)  # ‚úÖ Make sure directory exists
model_path = "tuning_results/ae_best_model_after_tuning.h5"  # ‚úÖ updated path

final_model, _ = train_autoencoder(
    X_train_0=X_trainval,
    X_val=X_val_scaled,
    input_dim=input_dim,
    encoding_dim=study.best_params['encoding_dim'],
    dropout=study.best_params['dropout'],
    noise_std=study.best_params['noise_std'],
    batch_size=study.best_params['batch_size'],
    epochs=epochs,
    model_path=model_path
)

# --- Final Evaluation on TEST set ---
metrics, ae_threshold = evaluate_autoencoder(final_model, X_test_scaled, y_test, visualize=True)

# --- Save Results ---
os.makedirs("tuning_results", exist_ok=True)
pd.DataFrame({k: [v] for k, v in metrics.items() if k != "report"}).to_csv("tuning_results/evaluation_summary.csv")
with open("tuning_results/classification_report.txt", "w") as f:
    f.write(metrics["report"])

styled_print("‚úÖ Results saved.")
````

### **2. Unit Testing** <br>

````python
import unittest

class TestAutoencoderTuning(unittest.TestCase):

    def test_model_training(self):
        """Check model is trained and returned properly."""
        model, history = train_autoencoder(
            X_train_0=X_train_0,
            X_val=X_val_scaled,
            input_dim=X_train_scaled.shape[1],
            encoding_dim=32,
            dropout=0.2,
            epochs=2,
            batch_size=128
        )
        self.assertTrue(hasattr(model, "summary"))
        self.assertGreater(len(history.history['loss']), 0)

    def test_evaluation_output(self):
        """Check that evaluate_autoencoder returns valid metrics."""
        model, _ = train_autoencoder(
            X_train_0=X_train_0,
            X_val=X_val_scaled,
            input_dim=X_train_scaled.shape[1],
            encoding_dim=32,
            dropout=0.2,
            epochs=2,
            batch_size=128
        )
        metrics, threshold = evaluate_autoencoder(
            model=model,
            X_data=X_test_scaled,
            y_data=y_test,
            visualize=False
        )
        self.assertIn("precision", metrics)
        self.assertIn("recall", metrics)
        self.assertIn("f1_score", metrics)
        self.assertIsInstance(threshold, float)

    def test_tuning_results_structure(self):
        """Ensure tuning results list is populated and structured."""
        self.assertIsInstance(tuning_results, list)
        if tuning_results:
            result = tuning_results[0]
            for key in ["encoding_dim", "dropout", "batch_size", "auc_roc", "pr_auc", "f1_score"]:
                self.assertIn(key, result)

# Run the tests
unittest.TextTestRunner().run(unittest.TestLoader().loadTestsFromTestCase(TestAutoencoderTuning))

````
